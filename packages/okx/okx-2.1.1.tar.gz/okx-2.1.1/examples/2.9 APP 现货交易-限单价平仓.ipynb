## 1. APP 现货交易-限单价平仓函数总览

|方法|解释|
|:---|:---|
|close_limit|限价单卖出|

## 2. 模型实例化

```python
from okx.app import OkxSPOT
from pprint import pprint
```

```python
# 现货交易账户需要秘钥
key = "xxx"
secret = "xxx"
passphrase = "xxx"
# 实盘：0，虚拟盘：1
flag = '0'
# 使用http和https代理，proxies={'http':'xxxxx','https:':'xxxxx'}，与requests中的proxies参数规则相同
proxies = {}
# 转发：需搭建转发服务器，可参考：https://github.com/pyted/okx_resender
proxy_host = None

okxSPOT = OkxSPOT(
    key=key, secret=secret, passphrase=passphrase, proxies=proxies, proxy_host=proxy_host,
)
trade = okxSPOT.trade
```

## 3. 同步 非堵塞 固定价格平仓（卖出）

```python
# block = False
close_limit3 = okxSPOT.trade.close_limit(
    instId='MANA-USDT',  # 产品
    closePrice=1,  # 平仓价格 closePrice 和 tpRate必须填写其中一个
    # tpRate=0.1,  # 挂单止盈率 以当前成交价为准
    quantity=10,  # 平仓数量
    block=False,  # 是否以堵塞的模式
    tag='',  # 订单标签
    clOrdId='',  # 客户自定义订单ID
    meta={},  # 向回调函数中传递的参数字典
)

pprint(close_limit3)
```

## 4. 同步 非堵塞 止盈价格平仓（卖出）

以当前价格作为基准，按照止盈率挂单卖出

```python
# 设置tpRate = 0.1，止盈率为10%
close_limit4 = okxSPOT.trade.close_limit(
    instId='MANA-USDT',  # 产品
    # closePrice=10,  # 平仓价格 closePrice 和 tpRate必须填写其中一个
    tpRate=0.1,  # 挂单止盈率
    quantity=10,  # 平仓数量
    block=False,  # 是否以堵塞的模式
    tag='',  # 订单标签
    clOrdId='',  # 客户自定义订单ID
    meta={},  # 向回调函数中传递的参数字典
)

pprint(close_limit4)
```

## 5. 同步 堵塞 止盈价格平仓（卖出） 超时取消

```python
# quantity = 'all' 表示全部可交易货币
# 设置block、timeout、cancel
close_limit5 = okxSPOT.trade.close_limit(
    instId='MANA-USDT',  # 产品
    # closePrice=10,  # 平仓价格 closePrice 和 tpRate必须填写其中一个
    tpRate=0.1,  # 挂单止盈率
    cancel=True,  # 未完全成交是否取消订单
    tag='',  # 订单标签
    clOrdId='',  # 客户自定义订单ID
    meta={},  # 向回调函数中传递的参数字典

    quantity='all',  # 平仓数量
    block=True,  # 是否堵塞
    timeout=5,  # 等待订单成功的超时时间
    delay=0.2,  # 检测订单状态的间隔 (秒)
)

pprint(close_limit5)
```

## 6. 同步 堵塞 止盈价格平仓（卖出） 回调函数

```python
# 执行成功回调
def callback6(information):
    print('callback')
    pprint(information)


# 执行错误回调
def errorback6(information):
    print('errorback')
    pprint(information)
```

```python
# 设置callback与errorback
close_limit6 = okxSPOT.trade.close_limit(
    instId='MANA-USDT',  # 产品
    # closePrice=1000,  # 平仓价格 closePrice 和 tpRate必须填写其中一个
    tpRate=0.1,  # 挂单止盈率
    quantity='all',  # 平仓数量
    block=True,  # 是否堵塞
    timeout=5,  # 等待订单成功的超时时间
    delay=0.2,  # 检测订单状态的间隔 (秒)
    cancel=True,  # 未完全成交是否取消订单
    tag='',  # 订单标签
    clOrdId='',  # 客户自定义订单ID
    meta={},  # 向回调函数中传递的参数字典

    callback=callback6,  # 开仓成功触发的回调函数
    errorback=errorback6,  # 开仓失败触发的回调函数
)
```

## 7. 异步 堵塞 止盈价格平仓（卖出） 回调函数

```python
# 执行成功回调
def callback7(information):
    print('thread callback')
    pprint(information)


# 执行错误回调
def errorback7(information):
    print('thread errorback')
    pprint(information)
```

```python
# 设置newThread=True
close_limit7 = okxSPOT.trade.close_limit(
    instId='MANA-USDT',  # 产品
    # closePrice=1000,  # 平仓价格 closePrice 和 tpRate必须填写其中一个
    tpRate=0.1,  # 挂单止盈率
    quantity='all',  # 平仓数量
    block=True,  # 是否堵塞
    timeout=5,  # 等待订单成功的超时时间
    delay=0.2,  # 检测订单状态的间隔 (秒)
    cancel=True,  # 未完全成交是否取消订单
    callback=callback7,  # 开仓成功触发的回调函数
    errorback=errorback7,  # 开仓失败触发的回调函数
    tag='',  # 订单标签
    clOrdId='',  # 客户自定义订单ID
    meta={},  # 向回调函数中传递的参数字典

    newThread=True,  # 是否开启一个新的线程维护这个订单
)
pprint(close_limit7)
print('-' * 30)
```
