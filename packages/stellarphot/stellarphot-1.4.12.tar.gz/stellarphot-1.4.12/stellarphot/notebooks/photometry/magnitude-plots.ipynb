```python
from astropy.table import Table
from matplotlib import pyplot as plt
import numpy as np
```

```python
tab = Table.read("some_name.csv")
```

```python
B = tab[tab["filter"] == "B"]
V = tab[tab["filter"] == "V"]
rp = tab[tab["filter"] == "rp"]
```

```python
plt.plot(B["mag_inst_cal"] - V["mag_inst_cal"], V["mag_inst_cal"], "o")
plt.grid()
```

```python
plt.plot(B["mag_inst_cal"], B["mag_cat"], ".")
plt.grid()
```

```python
plt.plot(B["mag_inst_cal"] - B["mag_cat"], ".")
plt.grid()
```
